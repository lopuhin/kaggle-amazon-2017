```python
import pickle
from matplotlib import pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
with open('../../matches_1k.pkl', 'rb') as f:
    all_matches = pickle.load(f)
```

```python
do_show = True
show_first = 20
shown = checked = 0
for path, matches in sorted(all_matches['right'].items()):
    checked += 1
    (best, best_idx), (second, _) = matches[:2]
    ratio = second / best
    if ratio > 1.5:
        if do_show:
            im_left = Image.open(path)
            right_path = all_matches['paths'][best_idx]
            print(path.split('/')[-1], right_path.split('/')[-1])
            im_right = Image.open(right_path)
            im = Image.new('RGB', (512, 256))
            im.paste(im_left, (0, 0))
            im.paste(im_right, (256, 0))
            plt.figure()
            plt.title('ratio {:.1f}, match L2 {:.0f}'.format(ratio, best))
            plt.imshow(im)
        shown += 1
        if shown == show_first:
            break
print('{:.0%} matches'.format(shown / checked))
```
